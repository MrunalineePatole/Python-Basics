```python
#Visualization Bar Chart,Scatter Plat & Normal Graph
from matplotlib import pyplot as pl

```

```python
import numpy as np
```

```python
x=np.arange(1,11)
```

```python
x
```

```python
y1=2*x
y2=4*x
```

```python
y1
y2

```

```python
pl.plot(x,y1,color='red',linewidth=3,linestyle=':')
pl.plot(x,y2,color='green',linewidth=3,linestyle=':')
pl.title("I M Graph")
pl.xlabel("X Axis")
pl.ylabel("Y Axis")
pl.grid(True)
pl.show()
```

```python
#Bar Chart
```

```python
student={'Anu':78,'Manu':89,'Ranu':90,'Chinu':67}
```

```python
student
```

```python
name=student.keys()
```

```python
Marks=student.values()
```

```python
name,Marks
```

```python
pl.bar(name,Marks,color='red')
pl.title("Bar Chart of Result")
pl.xlabel("Name of Student")
pl.ylabel("Marks Obtain")
pl.grid()
pl.show()
```

```python
#Scatter Plot
```

```python
x=[1,2,3,4,5,6]
y1=[10,20,30,40,50,60]
y2=[2,4,6,8,10,12]
```

```python
pl.scatter(x,y1)
pl.scatter(x,y2)
pl.title("I M Scatter Plot")
pl.xlabel("I M X Axis")
pl.ylabel("I M Y Axis")
pl.grid(True)
```
